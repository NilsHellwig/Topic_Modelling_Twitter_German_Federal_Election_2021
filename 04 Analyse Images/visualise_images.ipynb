```python
import matplotlib.pyplot as plt
import random
import glob
```

```python
num_images = 3
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def save_random_images(image_paths, save_path):
    fig, axs = plt.subplots(1, 3, figsize=(18, 6))
    for i, ax in enumerate(axs.flat):
        image = plt.imread(image_paths[i])
        ax.imshow(image, aspect='equal')
        ax.axis("off")

        # Rahmen hinzuf√ºgen
        rect = patches.Rectangle((0, 0), image.shape[1], image.shape[0], linewidth=2, edgecolor='black', facecolor='none')
        ax.add_patch(rect)

    plt.tight_layout()
    plt.savefig(save_path, format='jpeg')
```

```python
def get_random_image_paths(path, num_images):
    image_paths = []
    subfolders = glob.glob(path + "**/", recursive=True)
    for folder in subfolders:
        images = glob.glob(folder + "*.jpg")
        if images:
            selected_image = random.choice(images)
            image_paths.append(selected_image)
    return random.sample(image_paths, num_images)
```

```python
random_seed = 30 # 13  # 124 # 123 # 30
random.seed(random_seed)
path = "../Datasets/img_dataset_mentions/"
image_paths = get_random_image_paths(path, num_images)
image_paths = ['../Datasets/img_dataset_mentions/CDU_CSU/DerLenzMdB/1423962932833103883_0.jpg', '../Datasets/img_dataset_mentions/AFD/Alice_Weidel/1408871840215453700_0.jpg', '../Datasets/img_dataset_mentions/FDP/ria_schroeder/1374082373579595784_0.jpg']
print(image_paths)
save_random_images(image_paths, "mentions_examples.jpg")
```

```python
random_seed = 49
random.seed(random_seed)
path = "../Datasets/img_dataset_political_accounts/"
image_paths = get_random_image_paths(path, num_images)
print(image_paths)
save_random_images(image_paths, "politicial_accounts_examples.jpg")
```
